```python
import pandas as pd
```

```python
OLD_PAD = "C:/Users/Jerome/Dropbox/PC/Downloads/all_complexities_skeletonised_analysis.csv"
OLD_STRETCH = "C:/Users/Jerome/Dropbox/PC/Downloads/all_complexities_stretch_analysis.csv"
NEW = "C:/Users/Jerome/Documents/hanzi_data/complexities/all_complexities.csv"
```

```python
old_pad = pd.read_csv(OLD_PAD, index_col=0)
old_stretch = pd.read_csv(OLD_STRETCH, index_col=0)

new = pd.read_csv(NEW, index_col=0)
```

```python
old_pad = old_pad.rename({"original character": "original_character", "ID": "image_ID", "perimetric complexity": "old_perimetric_complexity", "pixel complexity": "old_pixel_complexity", "scale method": "scale_method"}, axis=1)
print(old_pad.shape)
print(old_pad.columns)
old_stretch = old_stretch.rename({"original character": "original_character", "ID": "image_ID", "perimetric complexity": "old_perimetric_complexity", "pixel complexity": "old_pixel_complexity", "scale method": "scale_method"}, axis=1)
print(old_stretch.shape)
print(old_stretch.columns)

old = pd.concat([old_pad, old_stretch], axis=0)
print(old.shape)
old.head()
```

```python
new = new.rename({"perimetric_complexity": "new_perimetric_complexity", "pixel_complexity": "new_pixel_complexity"}, axis=1)
print(new.shape)
new.head()
```

```python
df = old.merge(new, on=["original_character", "scale_method", "image_ID", "period"])
print(df.shape)
different = df[df["old_pixel_complexity"] != df["new_pixel_complexity"]][["original_character", "dataset_x", "period", "image_ID", "scale_method", "skeletonise method", "old_perimetric_complexity", "new_perimetric_complexity", "old_pixel_complexity", "new_pixel_complexity"]]
print(different.shape)
different.head()
```

```python
different.to_csv("unexplained_complexity_differences.csv")
```
